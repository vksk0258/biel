```python
import pandas as pd
import numpy as np

df = pd.read_csv("biel_train.csv")

# y_temp 결측치 제거
df = df.dropna()

# one-hot encoding (location)
df = pd.get_dummies(df)

columns = list(df.columns.values)
columns.remove('y_temp')

df
```

```python
from sklearn.model_selection import train_test_split

X = df[columns]
y = df[['y_temp']]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

mlr = LinearRegression()
mlr.fit(X_train, y_train)
```

```python
# x_test를 넣어 예측

y_predict = mlr.predict(X_test)
```

```python
my_data = [[48.8, 0.73, 4.4, 20, 2, 13, 14.0, 2.7, 20, 10, 27.5, 0, 1, 0, 0, 0, 0, 1, 0]]
my_predict = mlr.predict(my_data)
my_predict
```

```python
result = y_test[:]
result['y_predict'] = y_predict

result
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(16,12))
graph = fig.add_subplot(1,1,1)
graph.plot(result['y_temp'], marker='o', color='blue', label='real value')
graph.plot(result['y_predict'], marker='^', color='red', label='predicted value')
graph.set_title('Multiple Linear Regression(SW1)', size=20)
plt.xlabel('index', size=15)
plt.ylabel('y_temp', size=15)
plt.legend(loc='best')
```
